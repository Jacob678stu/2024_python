```python
import requests
from bs4 import BeautifulSoup

url = 'https://finance.yahoo.com/quote/NVDA' 

# 發送HTTP請求
response = requests.get(url)

# 用BeautifulSoup解析HTML
stock_soup = BeautifulSoup(response.text, 'html.parser')
print(stock_soup)
```

```python
price_div = stock_soup.find('div', {'class': 'container yf-1tejb6'})
print(price_div)
```
